### Variables Description

* **hotel** : Type of hotel, categorized as "Resort Hotel" or "City Hotel".

* **is_canceled** : Binary variable indicating if the booking was canceled (1 for canceled, 0 for not canceled).

* **lead_time** : Number of days between the booking date and the arrival date.

* **arrival_date_year** : The year of the arrival date.

* **arrival_date_month** : The month of the arrival date.

* **arrival_date_week_number** : The week number of the arrival date.

* **arrival_date_day_of_month** : The day of the arrival date.

* **stays_in_weekend_nights** : Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel.

* **stays_in_week_nights** : Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel.

* **adults** : Number of adults included in the booking.

* **children** : Number of children included in the booking.

* **babies** : Number of babies included in the booking.

* **meal** : Type of meal booked, categorized as "Undefined/SC" for no meal package, "BB" for Bed & Breakfast, "HB" for Half board, and "FB" for Full board.

* **country** : Country of origin of the guest.

* **market_segment** : Market segment designation, such as "Online TA" for Online Travel Agents and "Groups" for group bookings.

* **distribution_channel** : Booking distribution channel, including "TA" for Travel Agents, "Direct" for direct bookings, "Corporate" for corporate bookings, and "GDS" for Global Distribution Systems.

* **is_repeated_guest** : Binary variable indicating if the booking was from a repeated guest (1 for repeated guest, 0 for first-time guest).

* **previous_cancellations** : Number of previous booking cancellations by the guest.

* **previous_bookings_not_canceled** : Number of previous bookings not canceled by the guest.

* **reserved_room_type** : Code for the type of room reserved.

* **assigned_room_type** : Code for the type of room assigned to the booking.

* **booking_changes** : Number of changes or modifications made to the booking.
deposit_type: Type of deposit made for the booking, including "No Deposit," "Non Refund," and "Refundable."

* **deposit_type** : Type of deposit made for the booking, including "No Deposit," "Non Refund," and "Refundable."

* **agent** : ID of the travel agency that made the booking.

* **company** : ID of the company/entity that made the booking or is responsible for payment.

* **days_in_waiting_list** : Number of days the booking was in the waiting list before it was confirmed to the guest.

* **customer_type** : Type of booking, such as "Contract" for contract, "Group" for group, "Transient" for transient, and "Transient-Party" for transient party.

* **adr** : Average Daily Rate, representing the average rental income per paid occupied room in a given time period.

* **required_car_parking_spaces** : Number of car parking spaces required by the guest.
* **total_of_special_requests** : Number of special requests made by the guest.

* **reservation_status** : Last reservation status, including "Canceled," "Check-Out," and "No-Show."

* **reservation_status_date** : Date at which the last status was set.

The refined descriptions offer a comprehensive understanding of each variable, providing valuable insights for informed decision-making based on the data.

## 3. ***Data Wrangling***

````python
# duplicate values count
df.duplicated().sum()
````

````python
# Write your code to make your dataset analysis ready.
````

````python
# make a copy of the original dataset for removing duplicates and missing values
df_clear=df.copy()
````

````python
# Removing missing values
df_clear.dropna(inplace=True)
df_clear.isnull().sum()
````

````python
#remove the duplicate values frome the data set
df_clear.drop_duplicates(inplace=True)
df_clear.duplicated().sum()
````

````python
# the cleared dataset with no missing values and no duplicate values
df_clear
````

````python
# create another copy of the original dataset for further analysis
df_copy=df.copy()
````

````python

# Create a new 'arrival_date' column by combining day, month, and year, then convert it to datetime

df_copy['arrival_date'] = df_copy['arrival_date_day_of_month'].astype(str) + '-' + df_copy['arrival_date_month'] + '-' + df_copy['arrival_date_year'].astype(str)
df_copy['arrival_date']=pd.to_datetime(df_copy['arrival_date'])

# Extract the month from the 'arrival_date' column and count the bookings for each month

df_copy['arrival_date_month'] = df_copy['arrival_date'].dt.month
monthly_booking = df_copy['arrival_date_month'].value_counts().sort_index()
monthly_booking
````

````python
# Grouping the data by 'arrival_date_month' and calculating the mean of 'is_canceled' for each month
seasonal_cancellation=df_copy.groupby('arrival_date_month').agg({'is_canceled': 'mean'}).reset_index()

# Displaying the contents of the seasonal_cancellation dataframe
seasonal_cancellation
````

````python
# calculating the count of bookings for each hotel
booking_by_hotel=df_copy.groupby(['hotel']).size().reset_index(name='count')
booking_by_hotel
````

````python
# Calculating the count of bookings for each distribution channel
booking_counts_by_channel=df_copy.groupby('distribution_channel').size().reset_index(name='count')

# Display the resulting DataFrame
booking_counts_by_channel
````

````python
#calculating cancellation rate accross different distribution channel
cancellation_rate_by_channel=df_copy.groupby('distribution_channel')['is_canceled'].mean().reset_index(name='cancellation_rate')
cancellation_rate_by_channel
````

````python
# calculating total_no of stay for each booking
df_copy['total_stays'] = df_copy['stays_in_weekend_nights'] + df_copy['stays_in_week_nights']
df_copy['total_stays']
````

````python
# Calculating total ADR by distribution channel and hotel category
total_adr_by_channel = df_copy.groupby(['hotel','distribution_channel'])['adr'].sum().reset_index(name= 'Total ADR')
total_adr_by_channel
````

````python
# Calculating total revenue and adding it as a new column in the DataFrame
df_copy['revenue'] = df_copy['adr'] * df_copy['total_stays']

# Displaying the 'revenue' column
df_copy['revenue']
````

````python
#calculate which hotel produced more revenue
total_revenue_by_hotel = df_copy.groupby('hotel')['revenue'].sum().reset_index(name='Total Revenue')
total_revenue_by_hotel
````

````python
#calculating revenue acrros different hotel type and compare it
revenue_by_category = df_copy.groupby(['hotel', 'distribution_channel'])['revenue'].sum().reset_index()
revenue_by_category
````

````python
# Create a pivot table for visualization
pivot_table = revenue_by_category.pivot(index='hotel', columns='distribution_channel', values='revenue')

# Visualize Revenue Data using a Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title('Total ADR by Hotel and Distribution Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Hotel Type')
plt.show()
````

````python
# make a relation of adr with arrival month
average_adr_by_date = df_copy.groupby('arrival_date_month')['adr'].mean().reset_index(name = "Average ADR")
average_adr_by_date
````

````python
# Create a new column 'customer_type' based on the number of adults, children, and babies

df_copy['customer_type'] = "Undefined"  # Initialize as "Undefined" by default

# Define the conditions for different customer types
df_copy.loc[(df_copy['adults'] == 1) & (df_copy['children'] == 0) & (df_copy['babies'] == 0), 'customer_type'] = "Solo Adult"
df_copy.loc[(df_copy['adults'] > 1) & (df_copy['children'] == 0) & (df_copy['babies'] == 0), 'customer_type'] = "Adults"
df_copy.loc[(df_copy['adults'] > 0) & (df_copy['children'] > 0) & (df_copy['babies'] == 0), 'customer_type'] = "Family with Children"
df_copy.loc[(df_copy['adults'] > 0) & (df_copy['babies'] > 0), 'customer_type'] = "Family with Babies"
df_copy.loc[(df_copy['adults'] == 0) & (df_copy['children'] > 0) & (df_copy['babies'] > 0), 'customer_type'] = "Children with Babies"
df_copy.loc[(df_copy['adults'] == 0) & (df_copy['children'] > 0), 'customer_type'] = "Children"
df_copy.loc[(df_copy['adults'] == 0) & (df_copy['babies'] > 0), 'customer_type'] = "Babies"


# Display the updated DataFrame with the new 'customer_type' column




df_copy['customer_type'].value_counts()
````

````python
# Visualize the distribution of customer types
plt.figure(figsize=(8, 4))
sns.countplot(data=df_copy, x='customer_type')
plt.title('Distribution of Customer Types')
plt.xlabel('Customer Type')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
````
